## Transcripción con Whisper (large-v3)

```python
from faster_whisper import WhisperModel

model = WhisperModel(
    "large-v3",
    device="cuda",
    compute_type="float16"
)

segments, info = model.transcribe(
    audio_file,
    language="es",
    task="transcribe",        # explícito: transcribir, NO traducir
    beam_size=5,
    vad_filter=True,
    vad_parameters=dict(min_silence_duration_ms=500),
    condition_on_previous_text=False  # reduce alucinaciones/repeticiones en audios largos
)

segments_list = list(segments)

print("Idioma detectado:", info.language, "- probabilidad:", info.language_probability)
print("Segmentos:", len(segments_list))
```

```python
with open("transcripcion.txt", "w", encoding="utf-8") as f:
    for seg in segments_list:
        f.write(seg.text.strip() + "\n")

print("Transcripción guardada.")
```
